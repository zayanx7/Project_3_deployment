```python
import pandas as pd
import requests
import urllib.parse

# Define the function to get IMDb ID
def get_imdb_id(movie_title, api_key):
    base_url = "http://www.omdbapi.com/"
    params = {
        'apikey': api_key,
        't': urllib.parse.quote_plus(movie_title)
    }

    response = requests.get(base_url, params=params)
    print("Request URL:", response.url)
    
    data = response.json()

    if 'Error' in data:
        print(f"Error: {data['Error']}")
        return None

    imdb_id = data.get('imdbID', None)
    return imdb_id

# Replace 'YOUR_API_KEY' with your actual IMDb API key
api_key = '14bef856'

# Replace 'your_dataset.csv' with the actual path to your dataset
df = pd.read_csv(file_path)

# Create a new column 'IMDbID' to store IMDb IDs
df['IMDbID'] = df['Series_Title'].apply(lambda x: get_imdb_id(x, api_key))

# Display the DataFrame with IMDb IDs
print(df)

```
